Question 67

Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

```python
import datetime as dt

def fix(x) :
    if x.year >= 2061 :
        year = x.year - 100
    else :
        year = x.year

    return pd.to_datetime(dt.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix)
change_dt = df['Yr_Mo_Dy']
display(change_dt)
```

Question 76

년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
import datetime as dt

def change_date(x) :
    hour = x.split(':')[1]
    date = x.split(':')[0]

    if hour == '24' :
        hour = '00:00:00'
        final_date = pd.to_datetime(date + ' ' + hour) + dt.timedelta(days=1)

    else : 
        hour = hour + ':00:00'
        final_date = pd.to_datetime(date + ' ' + hour)
    return final_date

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

date_change = df
display(date_change)
```
